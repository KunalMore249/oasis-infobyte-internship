```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\\osisintership\\task1\\iris.csv')
df.head()
```

```python
df['Species'].value_counts()
```

```python
df.describe()
```

```python
df=df.drop(columns=['Id'])
df
```

```python
from sklearn.preprocessing import LabelEncoder
species_le=LabelEncoder().fit_transform(df['Species'])
print(species_le)
```

```python
df['Species']=species_le
df
```

```python
plt.scatter(df['PetalLengthCm'],df['Species'])
```

```python
plt.scatter(df['PetalWidthCm'],df['Species'],c='r')
plt.scatter(df['SepalLengthCm'], df['Species'],c='g')
plt.scatter(df['SepalWidthCm'], df['Species'], c='b')
plt.xlabel('Values')
plt.ylabel('Species')
plt.show()
```

```python
x=df.iloc[:,:-1].values
y=df['Species'].values
x,y
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.3,random_state=42)
x_train,x_test,y_train,y_test
```

```python
from sklearn.linear_model import LogisticRegression as LR
model=LR()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred,x_test
```

```python
from sklearn.metrics import accuracy_score as acs
score=acs(y_test,y_pred)
score
```

```python
a=model.predict(x)
b=df['Species'].values
print(a,b,sep='\n')
print(f"Accuracy={(acs(a,b)*100).__round__(2)}%")
```
